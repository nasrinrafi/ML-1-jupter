```python
import matplotlib.pyplot as plt
from random import randint as rnd

x=list(range(50))
y= [rnd(-10,10)for i in range (50)]
plt.plot(x,y)

plt.xlabel("time")
plt.ylabel("speed")           
```

```python
import matplotlib.pyplot as plt
from random import randint as rnd
x=list(range(10))
y= [rnd(-4,8)for i in range (10)]
z=[rnd(-4,4)for i in range (10)]
plt.plot(x,y,label="test battery")
plt.plot(x,z,label="tets1")
plt.legend()
plt.xlabel("time")
plt.ylabel("battery")
```

```python
import matplotlib.pyplot as plt
from random import randint as rnd
def r_random(num,min,max):
 point=[rnd(min,max)for i in range(num)]
 return point
x=r_random(20,0,100)
y=r_random(20,0,1000)
plt.xlabel("test")
plt.ylabel("number")
plt.title("value")
plt.scatter(x,y,c=r_random(20,0,30),cmap="magma",edgecolors="black")
plt.colorbar()
plt.show()
```
