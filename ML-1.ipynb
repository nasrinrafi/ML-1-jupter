```python
X=np.array(df.drop(['G3'],axis=1))
X
```

```python
Y=np.array(df['G3'])
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
Model=LinearRegression()
```

```python
Model.fit(x_train,y_train)
acc=Model.score(x_test,y_test)
print(acc)
```

```python
Model.intercept_
x_test[0]
y_test[0]
Model.coef_
Model.predict([x_test[0]])
```

```python
t=0
for i in zip(Model.coef_,x_test[0]):
 t+=i[0]*i[1]
 y=t+Model.intercept_
 print(y)
```

```python
prediction=Model.predict(x_test)
for i in range (len(y_test)):
    print(y_test.iloc[i],prediction[i],sep=">>")
```

```python
min_y_test=sum(y_test)/len(y_test)
rest=0
for yi,predi in zip (y_test,prediction):
 rest+=(yi-predi)**2
 print(rest)
tot=0
for yi in y_test:
    tot+=(yi-min_y_test)**2
    print(tot)
    r2=1-(rest/tot)
    print (r2)
    
```
